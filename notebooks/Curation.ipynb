```python
#swapping acq label with sub label
for root, dirs, files in os.walk("../output"):
    print(root)
    for filename in files:
        sub = re.search(r'(?<=acq-)[a-zA-Z0-9]+(?=_)', filename)
        acq = re.search(r'(?<=sub-)[a-zA-Z0-9]+(?=_)', filename)
        print(root+"/"+filename)
        src = root+"/"+filename
        if sub and acq:
            print(sub.group())
            print(acq.group())
            f = re.sub(r'(?<=sub-)[a-zA-Z0-9]+(?=_)', sub.group(), filename)
            f = re.sub(r'(?<=acq-)[a-zA-Z0-9]+(?=_)', acq.group(), f)
        print(root+"/"+f)
        dst = root+"/"+f
        os.rename(src, dst)
```
